### 2. Phân tích các feature theo nhóm:

```python
plt.figure(figsize= (15,15),dpi = 80)
sns.heatmap(df.corr(),cmap = 'Blues',annot = True)
```

```python
from scipy.stats import pearsonr
sns.set()
def test(df,lower_bound = -100000,upper_bound = 100000):
    c = df.corr().columns
    l = []
    for a in c:
        for b in c:
            if a!=b:
                if (b,a) not in l:
                    if (pearsonr(df[a],df[b])[0] >= lower_bound) & (pearsonr(df[a],df[b])[0] <= upper_bound):
                        print(f"{a} and {b} : {pearsonr(df[a],df[b])}")
                        l.append((a,b))
    return l
```

```python
def plot_test(df,a):
    fig, axes = plt.subplots(int(len(a)/2)+1,2,figsize = (18,6*(int(len(a)/2)+1)))
    col = 0
    row = 0
    for i in a:
        sns.regplot(x = df[i[0]],y = df[i[1]],ax = axes[row][col])
        if col == 0:
            col = 1
        else:
            row = row + 1
            col = 0
```

```python
a = test(df,0.8,1)
```

```python
plot_test(df,a)
```

```python
b = test(df,0.5,0.8)
```

```python
plot_test(df,b)
```

```python
c = test(df,0.3,0.5)
```

```python
plot_test(df,c)
```

```python
d = test(df,0.2,0.3)
```

```python
plot_test(df,d)
```
