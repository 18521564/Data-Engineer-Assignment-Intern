Q.4: We have an array of n elements A[1..n]. This array contains n different numbers from 0 to n. Given that there are totally n + 1 numbers from 0 to n, there is a missing number. (15 points) <br>
<br>
Example: <br>
+ array: [0, 1, 2, 4]. The missing number is 3.
+ array: [1, 4, 2, 3]. The missing number is 0.

```python
import random
import numpy as np
```

```python
a = np.delete(np.array([i for i in range(10)]),7)
```

```python
b = np.array(random.sample(range(10),9))
```

```python
def missing(array):
    n = sum([i  for i in range(len(array)+1)]) - sum(array)
    print("array: ",array,end=".")
    print(f"The missing is {n}")
```

```python
missing(a)
```

```python
missing(b)
```
