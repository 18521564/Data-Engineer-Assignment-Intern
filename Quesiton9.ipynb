Question 9: <br>
We have three identical six-sided dice. We roll one dice first and the remaining
two dice after that. What is the probability that the point obtained in the first roll is
greater than the sum of the points obtained in the second roll. 

Tóm tắt bài toán: <br>
Đổ 3 xúc xắc, 1 xúc xắc lần thứ nhất và 2 xúc xắc lần thứ hai. Tìm xác suất tổng điểm lần thứ nhất lớn hơn tổng điểm lần thứ hai. 

<b><i><em>Lời giải:</em></i></b> <br>
<em> Để đạt được yêu cầu bài toán, lần lượt các xúc xắc phải thỏa mãn các yêu cầu nhất định sau: </em> <br>

<b>Với xúc xắc thứ nhất:</b>Để xúc xắc đầu tiên thỏa điều kiện nó phải là 3 4 5 hoặc 6. Vì tối  nếu hai xúc sắc còn lại trong trường hợp tối thiểu ra 1 + 1 = 2. <br>
Ký hiệu: $$d1 \in {3,4,5,6}$$

<b>Với xúc sắc thứ hai </b>: Tùy thuộc vào giá trị của xúc xắc thứ nhất. Tập giá trị của xúc xắc thứ hai nằm trong khoảng từ 1 đến giá trị xúc xắc thứ nhất trừ đi 2 để đảm bảo nếu cộng tối thiểu 1 đơn vị nó vẫn không vượt quá giá trị của xúc xắc thứ nhất. 
Ký hiệu: $$ d2_i \in [1:d1_i - 2] $$
Ví dụ: Với d1 = 3 thì $d2 \in \{1\}$, với d1 = 4 thì $d2 \in \{1,2\}$

<b>Với xúc sắc thứ ba </b>: Tùy thuộc vào giá trị của xúc xắc thứ nhất và xúc xắc thứ hai. Có thể dễ dàng suy ra được số trường hợp thỏa mãn xúc xắc thứ ba từ xúc xắc thứ nhất và thứ hai chính là số đơn vị được phép thêm vào xúc xắc thứ hai sao cho giá trị không vượt quá xúc xắc thứ nhất. Lấy hiệu của xúc xắc thứ nhất trừ cho xúc xắc thứ hai trừ cho một (bảo đảm không vượt quá).
Ký hiệu: $$ d3_i \in [1:d2_i - d1_j - 1] $$
Ví dụ: <br> 
Với d1 = 3 thì d2 có duy nhất giá trị 1 và d2 chỉ có duy nhất giá trị 3 - 1 - 1 = 2 <br>
Với d1 = 4 thì $d2 \in \{1,2\}$. Với d1 = 4 và d2 = 1 thì tập d3  sẽ gồm 4-1-1 = 2 phần tử bao gồm {1,2} <br>

<em> Áp dụng vào tính toán: </em>
Xác xuất để ra được một mặt xúc xắc có điểm như ý muốn là 1/6. Để ra được ba mặt xúc xắc chính xác theo như ý muốn thì cần xác
suất $\frac{1}{6^3}$ <br>

Tổng số các trường hợp ba mặt chính xác thỏa yêu cầu bài toán sẽ được tính theo công thức sau: 
$$p = \frac{1}{6^3} \sum_{d_1=3}^{6} \ \sum_{d_2=1}^{d_1-2} (d_1 - (d_2 + 1))$$

$$p = 0.09259259259259259$$



```python
#sử dụng code để tính p:
count = 0 #tìm số trường hợp cả ba mặt xúc xắc thỏa mãn
for i in range(3, 7):
  for j in range(1, i-1):
    count += (i- j - 1)

p = count/(6**3)
print(p)
```
