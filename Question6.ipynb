Given the head of a sorted linked list, delete all duplicates such that each
element appears only once. Return the linked list sorted as well. (20 points)

+ head = [2, 2, 5] => [2, 5]
+ head = [1, 1, 3, 4, 4, 5, 6, 6] => [1, 3, 4, 5, 6]

<em><b><i> Lời giải: </i></b></em> <br>
Duyệt từng node một, nếu node tiếp theo của node hiện tại rỗng thì dừng thuật toán. <br>
Với mỗi node: <br>
1 Chừng nào node tiếp theo của node hiện tại vẫn chứa giá trị bằng giá trị node hiện tại, xóa node tiếp theo, node tiếp theo của node tiếp theo sẽ là node tiếp theo mới (node.next  = node.next.next). <br>
2 Di chuyển sang node tiếp theo.

```python
class Node:
    def __init__(self):
        self.data = None
        self.next = None
        

#linked list auto sort lúc insert:        
class linked_list:
    def __init__(self):
        self.head = None
   

    
    def add_node(self, num):
        #lấy iterator:
        temp = self.head
        
        #list rỗng:
        if temp is None:
            n = Node()
            n.data = num
            self.head = n
            return
        
        #tìm vị trí thích hợp để chèn (tìm node với data lớn hơn số cần chèn):
        
        #chèn ở head:
        if temp.data > num:
            n = Node()
            n.data = num
            n.next = temp
            self.head = n
            return
        
        #chèn khác chỗ head:
        while temp.next is not None:
            if temp.next.data > num:
                break
            temp = temp.next
        
        
        n = node()
        n.data = num
        n.next = temp.next
        temp.next = n
        return

    def list_print(self):
        if(self.head == None):
            print('list rong')
            return
        node = self.head 
        while node:
            print(node.data, end = " ")
            node = node.next
            
    def remove_dup(self):
        temp = self.head
        
        #list rỗng:
        if temp is None:
            print("list rỗng")
            return
        
        while temp.next is not None:
            while temp.data == temp.next.data:
                #lưu lại node next của next trước khi xóa:
                temp.next = temp.next.next
                if temp.next is None:
                    return
            temp = temp.next
        return
```

```python
a = [5,2,2]
llist = linked_list()
for i in a:
    llist.add_node(i)
```
